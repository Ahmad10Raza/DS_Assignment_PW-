### Q1. Which keyword is used to create a function? Create a function to re a list of odd numbers in the range of 1 to 25.

```python
def odd(num):
    l = []
    for i in range(1, num+1):
        if (i % 2 != 0):
            l.append(i)
    return l

num = 25
print(f"The odd numbers from 1 to {num} are {odd(num)}")
```

### Q2. Why *args and kwargs is used in some functions? Create a function each for "args and kwargs to demonstrate their use

<!-- args are used to return all the input arguments given to a function -->

```python
# '*args' is used to return all the input arguments passed to a function.
# '**kwargs' is used to return all the input arguments in dictionary's key and value pair form.

def ar(*args):
    return args


def kwar(**kwargs):
    return kwargs


argss = ar("A", 435, {"b": 535}, (342, 243, 53453), [5435, 5345, "Hi"])
print("*args implementation\n", argss, "\n")

kwargss = kwar(a=424, b=45345, c=["Aditya", "Hi"], d=(
    324, 23543, 3, 5424, 766))
print("**kwargs implementaion\n", kwargss)
```

### Q4. What is a generator function in python? Why yield keyword is used? Give an example of a generator function.

```python
# Genertor functions are used to generate objects that acts like an iterator.
# Yield keyword is used to return the value generated by the generated object.

# Example - Generate first 5 whole numbers.
def gen1(x):
    for i in range(x):
        yield i

gen_ob = gen1(5)
print(next(gen_ob))
print(next(gen_ob))
print(next(gen_ob))
print(next(gen_ob))
print(next(gen_ob))
```

### Q6. Write a python program to print the first 10 Fibonacci numbers using a while loop.

```python
def fibo(num):
    l = []
    i = 1
    sum = 0
    while (num):
        # print(sum)
        l.append(sum)
        j = i + sum
        i = sum
        sum = j
        num -= 1
    return l


num = 10
print(f"The first {num} numbers in fibonacci series are {fibo(10)}")
```
