## Q1. Write a python program to extract the video URL of the first five videos. 

```python
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument("--headless=new")
browser = webdriver.Chrome(
    options=options, service=Service(ChromeDriverManager().install()))

browser.get("https://www.youtube.com/@PW-Foundation/videos")

html = browser.page_source
soup = BeautifulSoup(html, "html.parser")

soupURL = soup.find("div", {"id": "contents"})

videoURL = (soupURL.find_all("a", {
    "class": "yt-simple-endpoint inline-block style-scope ytd-thumbnail"}))[0:5]

with open("Q1.csv", "w") as f:
    for i in videoURL:
        f.write("https://www.youtube.com"+str(i.get("href"))+"\n")

with open("Q1.csv") as f:
    print(f.read())
```

## Q2. Write a python program to extract the URL of the video thumbnails of the first five videos.

```python
from bs4 import BeautifulSoup
import time
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("--headless")
browser = webdriver.Chrome(
    options=options, service=Service(ChromeDriverManager().install()))

browser.get("https://www.youtube.com/@PW-Foundation/videos")

browser.execute_script("window.scrollTo(0,400)","")

html = browser.page_source
soup = BeautifulSoup(html, "html.parser")

videoThumb = soup.find_all("img",{"class":"yt-core-image--fill-parent-height"})[:5]

with open("Q2.csv", "w") as f:
    for i in videoThumb:
        f.write(str(i['src']+'\n'))
        
with open("Q2.csv") as f:
    print(f.read())
```

## Q3. Write a python program to extract the title of the first five videos.

```python
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver
options = webdriver.ChromeOptions()
options.add_argument("--headless=new")
browser = webdriver.Chrome(
    options=options, service=Service(ChromeDriverManager().install()))
browser.get("https://www.youtube.com/@PW-Foundation/videos")

html = browser.page_source

soup = BeautifulSoup(html, "html.parser")

soupTitle=soup.find("div",{"id": "contents"})

videoTitle = (soupTitle.find_all("a",{"class": "yt-simple-endpoint focus-on-expand style-scope ytd-rich-grid-media"}))[0:5]

with open("Q3.csv","w") as f:
    for i in videoTitle:
        f.write(str(i.get("title"))+"\n")

with open("Q3.csv") as f:
    print(f.read())
```

## Q4. Write a python program to extract the number of views of the first five videos.

```python
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("--headless=new")

browser = webdriver.Chrome(
    options=options, service=Service(ChromeDriverManager().install()))

browser.get("https://www.youtube.com/@PW-Foundation/videos")
html = browser.page_source
soup = BeautifulSoup(html, "html.parser")
soupNoOfViews=soup.find("div",{"id": "contents"})
videoNoOfViews = (soupNoOfViews.find_all("div",{"id": "metadata"}))[0:5]

with open("Q4.csv","w") as f:
     for i in videoNoOfViews:
        f.write(str(i.find_all("span")[1].text)+"\n")

with open("Q4.csv") as f:
    print(f.read())
```

## Q5. Write a python program to extract the time of posting of video for the first five videos.

```python
from bs4 import BeautifulSoup
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium import webdriver

options = webdriver.ChromeOptions()
options.add_argument("--headless=new")

browser = webdriver.Chrome(
    options=options, service=Service(ChromeDriverManager().install()))
browser.get("https://www.youtube.com/@PW-Foundation/videos")
html = browser.page_source
soup = BeautifulSoup(html, "html.parser")
soupTOP=soup.find("div",{"id": "contents"})
videoTOP = (soupTOP.find_all("div",{"id": "metadata"}))[0:5]

with open("Q5.csv","w") as f:
     for i in videoTOP:
        f.write(str(i.find_all("span")[2].text)+"\n")

with open("Q5.csv") as f:
    print(f.read())
```
