4. Check how many times a given number can be divided by 3 before it is less than or equal to 10.

```python
num = int(input("Enter a number"))
onum = num
count = int(0)
while (num > 10):
    num /= 3
    count += 1
print("The number", onum, "can be divided", count,
      "times by 3 before it's less than or equal to 10")
```

6. Use nested while loop to print 3 different pattern.

```python
# 1st pattern
i = int(5)
while (i > 0):
    j = int(0)
    while (j < i):
        print(j+1,end='')
        j+=1
    i -= 1
    print("")
```

```python
# 2nd Pattern
i = int(5)
count = int(1)
while (i > 0):
    j = i - count
    while (j < i):
        print(i - j, end='')
        j += 1
    i -= 1
    count += 1
    print("")
```

```python
# 3rd Pattern
i = int(5)
while (i > 0):
    j = i
    while (j > 0):
        print(j, end='')
        j -= 1
    print("")
    i -= 1
```
